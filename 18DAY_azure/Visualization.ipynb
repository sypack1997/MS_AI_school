### 바 그래프

```python
data = {
    '바나나' : 3, '사과' : 2, '오렌지' : 10
}
# 선:  plot, 바 : bar
plt.bar(data.keys(), data.values())
plt.title('과일 개수')
plt.xticks([0,1,2], data.keys())
plt.yticks(sorted(data.values()))
plt.grid(True)
plt.xlabel('과일')
plt.ylabel('개수')
plt.show()
```

```python
data = {
    '바나나' : 3, '사과' : 2, '오렌지' : 10
}
# 선:  plot, 바 : bar, 가로 바 : barh
plt.barh([0,1,2], data.values())
plt.title('과일 개수')
plt.yticks([0,1,2], data.keys())
plt.xticks(sorted(data.values()))
plt.grid(True)
plt.xlabel('개수')
plt.ylabel('과일')
plt.show()
```

### 파이차트

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Graphs']
explode = [0.05, 0.05, 0.05, 0.05]
colors = ['silver', 'gold', 'whitesmoke', 'lightgray']
# 파이차트 메서드 plt.pie

plt.pie(ratio, labels = labels, autopct = '%.1f%%', startangle = 260, counterclock = False, shadow = True, explode = explode, colors = colors)
plt.show()
```

```python
wedgeprops = {'width' : 0.7, 'edgecolor' : 'w', 'linewidth' : 5}
plt.pie(ratio, labels = labels, autopct = '%.1f%%', startangle = 260, counterclock = False, shadow = True, explode = explode, colors = colors, wedgeprops = wedgeprops)
```

```python
import numpy as np
from scipy import stats

#정규분포생성
data_a = np.random.normal(2.0, 2.0, 100)
```

```python
estimator = stats.gaussian_kde(data_a, bw_method = 'silverman')
x = np.arange(-10, 10)
K = estimator(x)
```

```python
plt.plot(x, K, label = 'gaussian_kde')
```

### Scatter Plot

```python
#seed는 난수 생성 시 키값을 통해 난수 고정
n = 50
np.random.seed(0)
x = np.random.rand(n)
y = np.random.rand(n)
```

```python
area = (30 * np.random.rand(n))**2
colors = np.random.rand(n)

plt.scatter(x,y, s = area, c = colors, alpha = 0.5, cmap = 'Spectral')
plt.colorbar()
plt.legend()
plt.show()
```

### 박스 그래프

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

plt.style.use('default')

data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3, 1.5, 1000)
data_c = np.random.normal(1.2, 1.5, 1000)

fig,ax = plt.subplots()
# 박스 그래프 : boxplot
# whis = 수염길이
ax.boxplot([data_a, data_b, data_c], notch = True, whis = 1.5, vert = True)
ax.set_ylim(-10.0, 10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')
plt.show()
```

### 바이올린 플롯(Violin plot)

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('default')

# 데이터를 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 1000)
data_c = np.random.normal(1.2, 1.5, 1000)

# 바이올린 : plt.violinplot
violin = plt.violinplot([data_a, data_b, data_c], showmeans = True, showextrema = True, showmedians = True)

plt.xticks(np.arange(1,4), labels = ['A', 'B', 'C'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('default')

# 데이터를 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 1000)
data_c = np.random.normal(1.2, 1.5, 1000)

# 바이올린 : plt.violinplot
violin = plt.violinplot([data_a, data_b, data_c], showmeans = True, showextrema = True, showmedians = True)

plt.xticks(np.arange(1,4), labels = ['A', 'B', 'C'])
violin['bodies'][0].set_facecolor('red')
violin['cbars'].set_edgecolor('orange')
violin['cmaxes'].set_edgecolor('black')
violin['cmins'].set_edgecolor('blue')
violin['cmeans'].set_edgecolor('green')
violin['cmedians'].set_edgecolor('purple')
plt.show()
```
